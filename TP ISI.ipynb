# Date

Separo la fecha de acuerdo a la estación del año (de Australia)

```python
def get_season(date):
    split_date = date.split(sep='-')
    month = int(split_date[1])
    if month==1 or month==2 or month==12:
        return 1
    elif month>=3 and month<=5:
        return 2
    elif month>=6 and month<=8:
        return 3
    return 4
    
```

```python
data['season']= data['Date'].apply(get_season)
```

```python
data = data.drop(['Date'], axis=1)
```

## Funciones de normalizacion

Este es uno de los métodos ampliamente utilizados para la normalización. El resultado de la normalización resta el valor mínimo de Dataframe y lo divide por la diferencia entre el valor más alto y el más bajo de la columna correspondiente.

```python
def minmax_norm(df):
    return (df - df.min()) / ( df.max() - df.min())
```

La normalización media es una de las formas más fáciles de normalizar los Dataframes de rangos variados. La normalización se hace restando la media y dividiendo por la desviación estándar para todos los elementos del Dataframe.

```python
def mean_norm(df):
    mean = df.mean()
    std = df.std()
    return df.apply(lambda x: (x-mean)/ std)
```

```python
selected_columns = data.loc[:,['MinTemp','MaxTemp','season']]
means = selected_columns.groupby('season').mean()
```

```python
data['MinTemp'] = data.apply(lambda x: x['MinTemp'] if not np.isnan(x['MinTemp']) else means['MinTemp'][x['season']], axis=1)
```

```python
data['MaxTemp'] = data.apply(lambda x: x['MaxTemp'] if not np.isnan(x['MaxTemp']) else means['MaxTemp'][x['season']], axis=1)
```
